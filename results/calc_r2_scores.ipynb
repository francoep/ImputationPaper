```python
import glob
import re
import sklearn.metrics
import scipy.stats
import numpy as np
import time
```

```python
def get_predictions(filenames):
    res=dict()
    for fname in filenames:
        with open(fname) as infile:
            for line in infile:
                if line.startswith('#'):
                    continue
                else:
                    items=line.rstrip().split()
                    pred_affinity=float(items[1])
                    true_affinity=abs(float(items[4]))
                    pose_score=float(items[0])
                    lig_filename=items[7]
                    m=re.search(r'(\S+)/...._._rec_...._(\S+)_(lig|uff_min)',lig_filename)
                    pocket=m.group(1)
                    ligand=m.group(2)
                    key=pocket+':'+ligand
                    if key not in res:
                        res[key] = (pred_affinity, true_affinity, pose_score)
                    else:
                        if res[key][2] < pose_score:
                            res[key] = (pred_affinity, true_affinity, pose_score)
        #print(res.keys())
        return list(res.values())
```

```python
dirnames=[x for x in glob.glob('it2_*/') if '_prog_med_' not in x]
```

```python
for d in dirnames:
    print(d)
    #start=time.time()
    for seed in range(5):
        with open(f'{d}s{seed}.r2.results','w') as outfile:
            todo=glob.glob(f'{d}*s{seed}n*_test.predictions')
            #print(todo)
            results=get_predictions(todo)
            preds=np.array([x[0] for x in results])
            true=np.array([x[1] for x in results])
    
            predictions=preds[true!=0]
            correctaff=true[true!=0]
            r2_score=sklearn.metrics.r2_score(correctaff,predictions)
            print('R2', r2_score)
            outfile.write(f'{r2_score}\n')
            #print('R', scipy.stats.pearsonr(correctaff,predictions)[0])
            #print('RMSE',np.sqrt(sklearn.metrics.mean_squared_error(correctaff,predictions)))
    #print('Took', time.time()-start)
```
